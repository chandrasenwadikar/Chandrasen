```python
# List of variables to map

varlist =  ['online_order', 'book_table']

# Defining the map function
def binary_map(x):
    return x.map({'Yes': 1, "No": 0})

# Applying the function to the housing list
df_rest[varlist] = df_rest[varlist].apply(binary_map)
```

```python
df_rest.head()
```

```python
# Creating a dummy variable for some of the categorical variables and dropping the first one.
dummy1 = pd.get_dummies(df_rest[['location', 'rest_type', 'dish_liked', 'cuisines','menu_item','listed_in(type)','listed_in(city)','reviews_list','rate']], drop_first=True)

# Adding the results to the master dataframe
df_rest = pd.concat([df_rest, dummy1], axis=1)
```

```python
# We have created dummies for the below variables, so we can drop them
df_rest = df_rest.drop(['location', 'rest_type', 'dish_liked', 'cuisines','menu_item','listed_in(type)','listed_in(city)','reviews_list','rate'], 1)
```

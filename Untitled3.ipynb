```python
paragraph = ["There was fox.",'It was brown in color.',"It was seen near lake."]
single_word_list =[]

for sentence in paragraph:
    for word in sentence.split():
        single_word_list.append(word)
print(single_word_list)


```

```python
single_word_list =[word for sentence in paragraph for word in sentence.split()]
print(single_word_list)
```

```python
vowels =['a','e','i','o','u']
vowels_from_sentence =[]
for sentence in paragraph:
    for word in sentence.split():
        if word[0].lower() in vowels:
            vowels_from_sentence.append(word)
            
print(vowels_from_sentence)
```

```python
vowels_comp =[word for sentence in paragraph for word in sentence.split()]
print(vowels_comp)
```

```python
students_data ={1:['Shivam Bansal',24],2:['Udit Bansal',25],3:['Sonam Kapoor',24]}
print(students_data)
```

```python
names_dict ={}

for roll_num, details in students_data.items():
    if roll_num%2 == 0:
        name_dict[roll_num]= details[0]
print(names_dict)
```

```python
names_comp = {roll_num : details[0] for roll_num, details in students_data.items() if roll_num%2==0}
print(names_comp)
```

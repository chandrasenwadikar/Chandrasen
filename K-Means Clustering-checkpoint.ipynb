## 3. Prepare the data for modeling

- R (Recency): Number of days since last purchase
- F (Frequency): Number of transactions 
- M (Monetary): Total amount of transactions(revenue contributed)

```python
#  Monetary 
grouped_df=retail_df.groupby('CustomerID')['amount'].sum()
grouped_df =grouped_df.reset_index()
grouped_df.head()
```

```python
# Frequency
frequency =retail_df.groupby('CustomerID')['InvoiceNo'].count()
frequency = frequency.reset_index()
frequency.columns = ['CustomerID','frequency']
frequency.head()
```

```python
# merge the two data frames

grouped_df = pd.merge(grouped_df, frequency , on='CustomerID', how='inner')
grouped_df.head()
```

```python
# Pre-requisite to recency 
# Convert to datetime

retail_df['InvoiceDate']= pd.to_datetime(retail_df['InvoiceDate'], format ='%d-%m-%Y %H:%M')

```

```python
retail_df.info()
```

```python
# compute the max date
max_date = max(retail_df['InvoiceDate'])
max_date
```

```python
# compute the difference
retail_df['diff'] = max_date-retail_df['InvoiceDate']
retail_df.head()
```

```python
# recency

last_purchase =retail_df.groupby('CustomerID')['diff'].min()
last_purchase = last_purchase.reset_index()
last_purchase.head()
```

```python
# Merge the data frame
grouped_df = pd.merge(grouped_df, last_purchase , on='CustomerID', how='inner')
#grouped_df.columns 
grouped_df.columns = ['CustomerID', 'amount', 'frequency','recency']
grouped_df.head()
```

```python
# number of days only
grouped_df['recency'] = grouped_df['recency'].dt.days
grouped_df.head()
```

```python
# Outlier treatment

plt.boxplot(grouped_df['recency'])
plt.show()
```

```python
# two ways to identify outlier
# - statistical
# - domain specific 
```

```python
# removing outliers (statistical method)
Q1 = grouped_df.amount.quantile(0.05)
Q3 = grouped_df.amount.quantile(0.95)
IQR = Q3 - Q1
grouped_df = grouped_df[(grouped_df.amount >= Q1 - 1.5*IQR) & (grouped_df.amount <= Q3 + 1.5*IQR)]

# outlier treatment for recency
Q1 = grouped_df.recency.quantile(0.05)
Q3 = grouped_df.recency.quantile(0.95)
IQR = Q3 - Q1
grouped_df = grouped_df[(grouped_df.recency >= Q1 - 1.5*IQR) & (grouped_df.recency <= Q3 + 1.5*IQR)]

# outlier treatment for frequency
Q1 = grouped_df.frequency.quantile(0.05)
Q3 = grouped_df.frequency.quantile(0.95)
IQR = Q3 - Q1
grouped_df = grouped_df[(grouped_df.frequency >= Q1 - 1.5*IQR) & (grouped_df.frequency <= Q3 + 1.5*IQR)]

```

```python
# Rescaling
rfm_df = grouped_df[['amount','frequency','recency']]
rfm_df.head()
```

```python
# Instantiate 

scaler = StandardScaler()

# fit_transform

rfm_df_scaled = scaler.fit_transform(rfm_df)
rfm_df_scaled.shape
```

### Hopkins Statistics

```python
from sklearn.neighbors import NearestNeighbors
from random import sample
from numpy.random import uniform
import numpy as np
from math import isnan
 
def hopkins(X):
    d = X.shape[1]
    #d = len(vars) # columns
    n = len(X) # rows
    m = int(0.1 * n) 
    nbrs = NearestNeighbors(n_neighbors=1).fit(X.values)
 
    rand_X = sample(range(0, n, 1), m)
 
    ujd = []
    wjd = []
    for j in range(0, m):
        u_dist, _ = nbrs.kneighbors(uniform(np.amin(X,axis=0),np.amax(X,axis=0),d).reshape(1, -1), 2, return_distance=True)
        ujd.append(u_dist[0][1])
        w_dist, _ = nbrs.kneighbors(X.iloc[rand_X[j]].values.reshape(1, -1), 2, return_distance=True)
        wjd.append(w_dist[0][1])
 
    H = sum(ujd) / (sum(ujd) + sum(wjd))
    if isnan(H):
        print(ujd, wjd)
        H = 0
 
    return H
```

```python
#First convert the numpy array that you have to a dataframe
rfm_df_scaled = pd.DataFrame(rfm_df_scaled)
rfm_df_scaled.columns = ['amount', 'frequency', 'recency']
```

```python
#Use the Hopkins Statistic function by passing the above dataframe as a paramter
hopkins(rfm_df_scaled)
```

```python
# Converting numpy array into data frame

rfm_df_scaled = pd.DataFrame(rfm_df_scaled)
rfm_df_scaled.columns = ['amount','frequency','recency']
rfm_df_scaled.head()
```

## 4. Modeling

```python
help(KMeans)
```

```python
# K-means with some arbitary k

kmeans = KMeans (n_clusters=4, max_iter=50)
kmeans.fit(rfm_df_scaled)
```

```python
kmeans.labels_
```

### Finding the optimal Number of Clusters

### SSD (Sum Squared Distance)

```python
# elbow-curve/SSD (Sum of Squared Distance)
ssd= [ ]
range_n_clusters = [2,3,4,5,6,7,8]
for num_clusters in range_n_clusters:
    kmeans = KMeans(n_clusters=num_clusters,max_iter=50)
    kmeans.fit(rfm_df_scaled)
    
    ssd.append(kmeans.inertia_)
    
# plot the SSDs for each n_clusters

plt.plot(ssd)
plt.show()
    
    
```

### Silhouette Analysis

```python
# data point num 10 (cluster-1)

# p = 100
# q = 10
# silhouette score = p-q/max(p,q)
# 90/100 = 0.9

# data point num 15 (cluster-1)

# p = 10
# q = 12
# -2/12= -1/6  ### this result is poor silhouette score

range_n_clusters = [2,3,4,5,6,7,8]
for num_clusters in range_n_clusters:
    # intialise kmeans
    kmeans = KMeans(n_clusters=num_clusters,max_iter=50)
    kmeans.fit(rfm_df_scaled)
    
    cluster_labels = kmeans.labels_
    
    # silhouette score
    silhouetter_avg = silhouette_score(rfm_df_scaled,cluster_labels)
    print("For n_clusters={0}, the silhouetter score is {1}".format(num_clusters,silhouetter_avg))
    
```

```python
# Final model with k =3
# K-means with some arbitary k=3

kmeans = KMeans (n_clusters=3, max_iter=50)
kmeans.fit(rfm_df_scaled)
```

```python
kmeans.labels_
```

```python
# assign the label

grouped_df['cluster_id'] = kmeans.labels_
grouped_df.head()
```

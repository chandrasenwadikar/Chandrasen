```python
titanic_df = titanic_df.drop(['Cabin'], axis=1)
titanic_df = titanic_df.dropna()

titanic_df.count()
```

```python
""" Data preprocessing function"""
def preprocess_titanic_df(df):
    processed_df = df.copy()
    le = preprocessing.LabelEncoder()
    processed_df.Sex = le.fit_transform(processed_df.Sex)
    processed_df.Embarked = le.fit_transform(processed_df.Embarked)
    processed_df = processed_df.drop(['PassengerId','Name','Ticket'],axis=1)
    return processed_df


processed_df = preprocess_titanic_df(titanic_df)
processed_df
```

```python
## splitting dataset into train and test
from sklearn.model_selection import train_test_split

X = processed_df.drop(['Survived'], axis=1).values # Features dataset
Y = processed_df['Survived'].values # Target variable

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=.2, random_state=42)
```

```python
## fitting the model - Linear Logistic Regression
from sklearn import linear_model

h = .02  # step size in the mesh
logreg = linear_model.LogisticRegression(C=1e5)
logreg.fit(X, Y)

y_predicted = logreg.predict(X_test)

from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_predicted)
print("\n\nAccuracy:", accuracy*100,'%')

df_confusion = pd.crosstab(y_test, y_predicted, rownames=['Actual'],
                           colnames=['Predicted'], margins=True)
print("\n\nConfusion Matrix:\n", df_confusion)
```

```python
## fitting the model - Decision Tree
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier(max_depth=5)
dt.fit(X, Y)

y_predicted = dt.predict(X_test)

accuracy = accuracy_score(y_test, y_predicted)
print("\n\nAccuracy:", accuracy*100,'%')

df_confusion = pd.crosstab(y_test, y_predicted, rownames=['Actual'],
                           colnames=['Predicted'], margins=True)
print("\n\nConfusion Matrix:\n", df_confusion)
```

```python
## fitting the model - Random Forest
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(max_depth=5, n_estimators=100, max_features=5)
rf.fit(X, Y)


y_predicted = rf.predict(X_test)

accuracy = accuracy_score(y_test, y_predicted)
print("\n\nAccuracy:", accuracy*100,'%')

df_confusion = pd.crosstab(y_test, y_predicted, rownames=['Actual'],
                           colnames=['Predicted'], margins=True)
print("\n\nConfusion Matrix:\n", df_confusion)
```

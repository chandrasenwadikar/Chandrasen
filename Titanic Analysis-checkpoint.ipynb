Lets impute it, but how?
We will use CLass and age relation to impute.

```python
#now we will use the mean value of each class as imputed value
titanic_df.groupby(['Pclass'])['Age'].mean()
```

```python
def age_imputer(cols):
    age = cols[0]
    pclass = cols[1]
    
    if pd.isnull(age):
        if pclass == 1:
            return 38
        elif pclass == 2:
            return 29
        else:
            return 25
    else:
        return age
```

```python
titanic_df['Age'] = titanic_df[['Age', 'Pclass']].apply(age_imputer, axis=1)
titanic_df.isnull().sum()
```

```python
## dropping records with null value (2 rec only now)
titanic_df.dropna(inplace=True)
titanic_df.isnull().sum()
```

Converting categorical variables to a dummy indicators

```python
gender = pd.get_dummies(titanic_df['Sex'],drop_first=True)
gender.head()
```

```python
embark_location = pd.get_dummies(titanic_df['Embarked'],drop_first=True)
embark_location.head()
```

```python
titanic_df.head()
```

```python
titanic_df.drop(['Sex', 'Embarked'],axis=1,inplace=True)
titanic_dummy = pd.concat([titanic_df,gender,embark_location],axis=1)
titanic_dummy.head()
```

```python
#Checking for independence between features
sb.heatmap(titanic_dummy.corr())
```

```python
#Fare and Pclass are not independent of each other, so I am going to drop these.
titanic_dummy.drop(['Fare', 'Pclass'],axis=1,inplace=True)
titanic_dummy.head()
```

```python
# remove passengerid, name and ticket
titanic_cleaned = titanic_dummy.drop(['PassengerId','Name','Ticket'], axis=1)
titanic_cleaned.head()
```

```python
Y_var = 'Survived'
X_var = list(titanic_cleaned)
X_var.remove(Y_var)
X_var
```

```python
X = titanic_cleaned[X_var]
X.head()
```

```python
y = titanic_cleaned[Y_var]
y.head()
```

```python
## splitting into train and test sets
from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = .2, random_state=25)
```

# Model evaluation

```python
## fitting the model - Linear Logistic Regression
print('Logistic Regression')
from sklearn import linear_model

logreg = linear_model.LogisticRegression(C=1e5)
logreg.fit(X, y)

y_predicted = logreg.predict(X_test)

from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_predicted)
print("\n\nAccuracy:", accuracy*100,'%')

df_confusion = pd.crosstab(y_test, y_predicted, rownames=['Actual'],
                           colnames=['Predicted'], margins=True)
print("\n\nConfusion Matrix:\n", df_confusion)
```
